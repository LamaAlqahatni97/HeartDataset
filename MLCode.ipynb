```python
import pandas as pd
df=pd.read_csv("cleveland (1).csv",sep=",")
df
```

```python
df.columns = ['age', 'sex', 'cp', 'trestbps', 'chol',

              'fbs', 'restecg', 'thalach', 'exang', 

              'oldpeak', 'slope', 'ca', 'thal', 'target']
df
```

```python
modifiedFlights = df.dropna()
modifiedFlights.isnull().sum()
df=df.dropna()
df
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score


import warnings
warnings.filterwarnings('ignore')
```

```python
df['target'].replace([0],['0'],inplace=True)
df['target'].replace([1],['1'],inplace=True)
df['target'].replace([2],['1'],inplace=True)
df['target'].replace([3],['1'],inplace=True)
df['target'].replace([4],['1'],inplace=True)
df['sex'] = df.sex.map({0: 'female', 1: 'male'})
df['sex'] = df.sex.map({'female': 0, 'male': 1})
df
```

```python
del df['age']
del df['trestbps']
del df['chol']
del df['thalach']
df
```

```python
# Prepare data for modeling

y = df.target
X = df.drop('target', axis=1)

# setting up testing and training sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=27)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
classifier = DecisionTreeClassifier()



criterion = [ 'gini', 'entropy']
max_depth = [1,3,5 , None]
splitter = [ 'best', 'random']

grid = GridSearchCV(estimator=classifier, cv=3, param_grid=dict(criterion=criterion,max_depth= max_depth,splitter=splitter))

grid.fit(X,y)

print("Best Score via GridSearch", grid.best_score_)
print("Best parameters via GridSearch", grid.best_params_)


#Predict values based on new parameters
y_pred_acc = grid.predict(X_test)
# New Model Evaluation metrics 
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score

print('Accuracy Score : '  + str(accuracy_score(y_test,y_pred_acc)))
print('Precision Score : ' + str(precision_score(y_test, y_pred_acc, pos_label='1')))
print('Recall Score : '    + str(recall_score(y_test,y_pred_acc, pos_label='1')))
print('F1 Score : '        + str(f1_score(y_test,y_pred_acc, pos_label='1')))

# SVR (Grid Search) Confusion matrix
from sklearn.metrics import confusion_matrix
conf = confusion_matrix(y_test, y_pred_acc)
print('Confusion Matrix : \n' + str(conf))


TP = conf[0,0]
FP = conf[0,1]
TN = conf[1,0]
FN = conf[1,1]

sensitivity = TP / float(FN + TP)

print(' sensitivity : \n' + str(sensitivity))

specificity = TN / (TN + FP)

print(' specificity : \n' + str(specificity))
```

```python

from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn.linear_model import LogisticRegression

from sklearn import metrics
from sklearn.model_selection import KFold
KFold
# prepare cross validation
kfold = KFold(n_splits=10, random_state=123, shuffle=True)

# enumerate splits on the features (X) set 
for train_indx, test_indx in kfold.split(X):    
    X_train = X.iloc[train_indx]                  # -- set of X_train from X-set
    X_test  = X.iloc[test_indx]                   # -- set of X_test from X-set
    y_train, y_test = y.iloc[train_indx], y.iloc[test_indx] # -- set of y_train, y_test from y-set
    
    # Apply these set as a regular process for any classifier to test
    clf = LogisticRegression().fit(X_train,y_train)
    y_pred = clf.predict(X_test)
    
    # Model Evaluation metrics 
    print('Accuracy Score : '  + str(accuracy_score(y_test,y_pred)))
    print('Precision Score : ' + str(precision_score(y_test,y_pred, pos_label='1')))
    print('Recall Score : '    + str(recall_score(y_test,y_pred, pos_label='1')))
    print('F1 Score : '        + str(f1_score(y_test,y_pred, pos_label='1')))

    #Logistic Regression Classifier Confusion matrix
    print('Confusion Matrix : \n' + str(confusion_matrix(y_test,y_pred)))
    conf = confusion_matrix(y_test, y_pred)

TP = conf[0,0]
FP = conf[0,1]
TN = conf[1,0]
FN = conf[1,1]

sensitivity = TP / float(FN + TP)

print(' sensitivity : \n' + str(sensitivity))

specificity = TN / (TN + FP)

print(' specificity : \n' + str(specificity))
```
